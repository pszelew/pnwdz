```python
offers_df = pd.read_csv('oferta_eksport_2022-01-05_11_43_51.csv', sep=';')
```

```python
def split_address(location):
    address = []
    split = location.split(', ')
    for s in split:
        if not 'powiat' in s:
            address.append(s)
        else:
            break
    return ', '.join(address)
```

```python
offers_df.columns = ['position', 'location', 'contract', 'employer', 'starting_from', 'err']
offers_df = offers_df.drop('err', axis=1)
offers_df['location'] = offers_df['location'].apply(lambda x: x.replace('"', ''))
offers_df['contract'] = offers_df['contract'].apply(lambda x: x.replace('"', ''))
offers_df['employer'] = offers_df['employer'].apply(lambda x: x.replace('"', ''))
offers_df['address'] = offers_df['location'].apply(lambda x: x.split(', ')[0])
# offers_df['open_street_raw'] = offers_df['address'].apply(lambda x: geocoder.osm(x).json['raw'])
offers_df.to_csv('oferta_eksport.csv', index=False)
```

### Usuwanie błędów, długie i monotonne

```python
def get_raw_geo_data(row):
    g = geocoder.osm(row['address'])
    if g.json is None:
        raw = 'Not Found'
    else:
        raw = g.json['raw']
    return raw
```

```python
offers_df['address'] = offers_df.apply(lambda row: ', '.join(row['location'].split(', ')[:2]) if row['open_street_raw'] == 'Not Found' else row['address'], axis=1)
```

```python
offers_df['open_street_raw'] = offers_df.apply(lambda row: get_raw_geo_data(row) if row['open_street_raw'] == 'Not Found' else row['open_street_raw'], axis=1)
```

```python
offers_df['address'] = offers_df.apply(lambda row: row['location'].split(', ')[1] if row['open_street_raw'] == 'Not Found' else row['address'], axis=1)
```

```python
offers_df['open_street_raw'] = offers_df.apply(lambda row: get_raw_geo_data(row) if row['open_street_raw'] == 'Not Found' else row['open_street_raw'], axis=1)
```

```python
offers_df = offers_df[offers_df['open_street_raw'] != 'Not Found']
```
